In the field of Data Science, data is often formatted as a comma-delimited (CSV) file, where each line in a file corresponds to a field's value. Refer to Chapter 9 (p. 106) for more information on comma-delimited files.

The file contains statistics about housing in the city of Denton from 2008 to 2014. Write a program to implement the following questions.

(a) (5 pts) Calculate the difference in the number of Occupied Housing Units from year to year and print it. The difference must be calculated for the consecutive years such as 2008-2009, 2009-2010 etc. Finally, print the values in the ascending order.

(b) (5 pts) For all the years, calculate the percentage of housing units which are vacant an occupied. Print the results in the following format:

Year Vacant Housing Units Occupied Housing Units

    2008  30%   70%                          
	2009 ----- -----
	2010 ----- -----
    2011 ----- -----
	2012 ----- -----
    2013 ----- -----
	2014 ----- -----
  
  

(c) (5 pts) Calculate and print the valued and years in which the highest number of housing units were vacant and occupied. Print the results in the following format:  

                              Year	Value
	Vacant Housing Units	  -----   -----
	Occupied Housing Units	-----   -----

(d) (5 pts) Calculate the harmonic mean of the total housing units and print it out. You can find the information about harmonic mean here: https://ncalculators.com/statistics/harmonic-mean-calculator.htm

```python
import pandas as pd
from scipy.stats import hmean   #import scipy.stats for harmonic mean calculation.

# Loading the CSV file using read_csv.
loc_file = '/assignment1_denton_housing.csv'
data = pd.read_csv(loc_file)


occupied_housing = data[data['title_field'] == 'Occupied Housing Units'].sort_values(by='year')
occupied_housing['difference'] = occupied_housing['value'].diff()
occupied_housing = occupied_housing.dropna(subset=['difference'])
sorted_differences = occupied_housing.sort_values(by='difference')

print("(a) Difference in the number occupied housing units by years")
for index, row in sorted_differences.iterrows():
    previous_year = row['year'] - 1
    print(f"{previous_year}-{int(row['year'])}: {int(row['difference'])} units")

# (b) Calculate the percentage of housing units which are vacant and occupied for all years
vacant_housing = data[data['title_field'] == 'Vacant Housing Units']
total_housing = data[data['title_field'] == 'Total Housing Units']

print("\n(b) Vacant and Occupied Housing Units:")


for year in total_housing['year'].unique():
    total = total_housing[total_housing['year'] == year]['value'].values[0]
    vacant = vacant_housing[vacant_housing['year'] == year]['value'].values[0]
    occupied = occupied_housing[occupied_housing['year'] == year]['value'].values[0] if not occupied_housing[occupied_housing['year'] == year].empty else 0
    vacant_pct = (vacant / total) * 100
    occupied_pct = (occupied / total) * 100
    print(f"{year} {vacant_pct:.2f}% {occupied_pct:.2f}%")

# (c) Calculate and print the years with the highest number of housing units vacant and occupied
maximum_vacant = vacant_housing.loc[vacant_housing['value'].idxmax()]
maximum_occupied = occupied_housing.loc[occupied_housing['value'].idxmax()]

print("\n(c) Year and Value with Highest Number of Housing Units:")
print("                          Year    Value")
print(f"Vacant Housing Units      {int(maximum_vacant['year'])}   {int(maximum_vacant['value'])}")
print(f"Occupied Housing Units    {int(maximum_occupied['year'])}   {int(maximum_occupied['value'])}")

# (d) Calculating harmonic mean using scipy package.
total_units = total_housing['value'].tolist()
harmonic_mean = hmean(total_units)

print("\n(d) Harmonic Mean of Total Housing Units:")
print(f"Harmonic Mean: {harmonic_mean:.2f}")
```

```python
import csv
import os
import uuid

# Constants
FLIGHT_FILE = "flights.csv"
RESERVATION_FILE = "reservations.csv"

# Initialize CSV files
def initialize_files():
    '''
    Prompt:
    "Generate a Python function named `initialize_files` that checks if two CSV files exist: `flights.csv` and `reservations.csv`.
    If either file does not exist, the function should create them with appropriate headers.
    The `flights.csv` should have headers 'FlightID', 'Origin', 'Destination', 'Seats', 'AvailableSeats', 'Price',
    while `reservations.csv` should have headers 'ReservationID', 'FlightID', and 'PassengerName'."

    Reverse Prompt:
    "The `initialize_files` function checks if `flights.csv` and `reservations.csv` exist.
    If they don't, it creates these files with the required headers for storing flight and reservation data."

    Edge Case:
    - Permission errors: Ensure that the function handles permission issues where the files cannot be created or modified.
    - File already exists: The function should not overwrite existing files, it should only create them if they are missing.
    '''
    if not os.path.exists(FLIGHT_FILE):
        with open(FLIGHT_FILE, mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(["FlightID", "Origin", "Destination", "Seats", "AvailableSeats", "Price"])
    if not os.path.exists(RESERVATION_FILE):
        with open(RESERVATION_FILE, mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(["ReservationID", "FlightID", "PassengerName"])


# Function to add new flights
def add_flight():
    '''
    Prompt:
    "Write a Python function named `add_flight` that prompts the user to input flight details: 'Flight ID',
    'Origin', 'Destination', 'Total Seats', and 'Ticket Price'. It should append this information to the `flights.csv` file."

    Reverse Prompt:
    "The `add_flight` function prompts the user for flight details and appends the new flight information to the `flights.csv` file."

    Edge Case:
    - Invalid input: Ensure that the 'Total Seats' and 'Price' are numeric. If invalid, the user should be prompted to re-enter valid values.
    - File write errors: Handle cases where the file cannot be written due to permission or other issues.
    '''
    flight_id = input("Enter Flight ID: ")
    origin = input("Enter Origin: ")
    destination = input("Enter Destination: ")
    seats = int(input("Enter Total Seats: "))
    price = float(input("Enter Ticket Price: "))
    try:
        with open(FLIGHT_FILE, mode='a', newline='') as file:
            writer = csv.writer(file)
            writer.writerow([flight_id, origin, destination, seats, seats, price])
        print("Flight added successfully!")
    except Exception as e:
        print(f"Error adding flight: {e}")


# Function to find flights
def find_flight():
    '''
    Prompt:
    "Create a Python function named `find_flight` that allows the user to search for available flights by 'Origin' and 'Destination'.
    The function should display flight information such as 'Flight ID', 'Available Seats', and 'Price' if matching flights are found."

    Reverse Prompt:
    "The `find_flight` function searches for flights based on 'Origin' and 'Destination' input by the user
    and displays relevant flight details like flight ID, available seats, and price."

    Edge Case:
    - No results: If no flights match the given search criteria, the user should be informed that no flights were found.
    - Case sensitivity: The function should handle case-insensitive comparisons for origin and destination.
    '''
    search_origin = input("Enter Origin: ")
    search_destination = input("Enter Destination: ")
    try:
        with open(FLIGHT_FILE, mode='r') as file:
            reader = csv.reader(file)
            next(reader)
            results = [row for row in reader if row[1].lower() == search_origin.lower() and row[2].lower() == search_destination.lower()]
            if results:
                for flight in results:
                    print(f"FlightID: {flight[0]}, Available Seats: {flight[4]}, Price: ${flight[5]}")
            else:
                print("No flights found!")
    except Exception as e:
        print(f"Error finding flights: {e}")


# Function to book a seat
def book_seat():
    '''
    Prompt:
    "Write a Python function named `book_seat` that allows the user to book a seat on a flight by providing 'Flight ID' and 'Passenger Name'.
    The function should check if the flight has available seats, update the seat count in `flights.csv`, and store the reservation
    in `reservations.csv` with a unique Reservation ID."

    Reverse Prompt:
    "The `book_seat` function checks if a flight has available seats, books a seat, updates the seat count in `flights.csv`,
    and creates a reservation in `reservations.csv`."

    Edge Case:
    - Flight unavailable: If the flight is fully booked or does not exist, the user should be informed.
    - Reservation creation failure: Handle any file write errors when updating the reservation details.
    '''
    flight_id = input("Enter Flight ID: ")
    passenger_name = input("Enter Passenger Name: ")
    reservation_id = str(uuid.uuid4())[:8]  # Generate a unique reservation ID
    try:
        rows = []
        booked = False
        with open(FLIGHT_FILE, mode='r') as file:
            reader = csv.reader(file)
            for row in reader:
                if row[0] == flight_id and int(row[4]) > 0:
                    row[4] = str(int(row[4]) - 1)
                    booked = True
                rows.append(row)

        if booked:
            with open(FLIGHT_FILE, mode='w', newline='') as file:
                writer = csv.writer(file)
                writer.writerows(rows)
            with open(RESERVATION_FILE, mode='a', newline='') as file:
                writer = csv.writer(file)
                writer.writerow([reservation_id, flight_id, passenger_name])
            print(f"Seat booked successfully! Your Reservation ID is: {reservation_id}")
        else:
            print("Flight not available or fully booked!")
    except Exception as e:
        print(f"Error booking seat: {e}")


# Function to cancel a reservation
def cancel_reservation():
    '''
    Prompt:
    "Write a Python function named `cancel_reservation` that allows the user to cancel a reservation by providing the 'Reservation ID'.
    The function should search for the reservation in `reservations.csv`, remove it, and update the seat count for the corresponding flight."

    Reverse Prompt:
    "The `cancel_reservation` function cancels a reservation by finding the Reservation ID, deleting the record from `reservations.csv`,
    and updating the seat count for the corresponding flight."

    Edge Case:
    - Reservation not found: If the reservation ID does not exist, inform the user that no reservation was canceled.
    - File write issues: Handle errors during file update or reservation deletion.
    '''
    reservation_id = input("Enter Reservation ID: ")
    try:
        rows = []
        flight_id = ""
        canceled = False
        with open(RESERVATION_FILE, mode='r') as file:
            reader = csv.reader(file)
            for row in reader:
                if row[0] == reservation_id:
                    flight_id = row[1]
                    canceled = True
                else:
                    rows.append(row)

        if canceled:
            with open(RESERVATION_FILE, mode='w', newline='') as file:
                writer = csv.writer(file)
                writer.writerows(rows)
            update_seat_count(flight_id, increase=True)
            print("Reservation cancelled successfully!")
        else:
            print("Reservation ID not found!")
    except Exception as e:
        print(f"Error canceling reservation: {e}")


# Function to update seat count
def update_seat_count(flight_id, increase=False):
    '''
    Prompt:
    "Write a Python function named `update_seat_count` that updates the seat count for a given flight.
    The function should either increase or decrease the number of available seats based on the `increase` flag,
    and save the changes in the `flights.csv` file."

    Reverse Prompt:
    "The `update_seat_count` function adjusts the available seat count for a flight by either increasing or decreasing it and saves the change in `flights.csv`."

    Edge Case:
    - Invalid flight ID: Handle the case where the flight ID does not exist in the `flights.csv` file.
    - File write issues: Ensure that any errors related to file writing are caught and reported to the user.
    '''
    try:
        rows = []
        with open(FLIGHT_FILE, mode='r') as file:
            reader = csv.reader(file)
            for row in reader:
                if row[0] == flight_id:
                    row[4] = str(int(row[4]) + 1 if increase else int(row[4]) - 1)
                rows.append(row)
        with open(FLIGHT_FILE, mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerows(rows)
    except Exception as e:
        print(f"Error updating seat count: {e}")


# Function to generate flight performance reports
def generate_report():
    '''
    Prompt:
    "Write a Python function named `generate_report` that reads the `flights.csv` file,
    calculates the number of tickets sold and the revenue for each flight, and then generates a performance report."

    Reverse Prompt:
    "The `generate_report` function generates a report showing the number of tickets sold and the revenue for each flight in `flights.csv`."

    Edge Case:
    - No flights available: If no data is available in the `flights.csv`, the report should indicate this.
    - File read issues: Handle cases where there might be errors reading the CSV file.
    '''
    try:
        with open(FLIGHT_FILE, mode='r') as file:
            reader = csv.reader(file)
            next(reader)
            print("\n--- Flight Performance Report ---")
            for row in reader:
                sold_tickets = int(row[3]) - int(row[4])
                revenue = sold_tickets * float(row[5])
                print(f"FlightID: {row[0]}, Tickets Sold: {sold_tickets}, Revenue: ${revenue:.2f}")
    except Exception as e:
        print(f"Error generating report: {e}")

# Initialize files
initialize_files()

# Menu-driven interface
while True:
    print("\nFlight Booking System")
    print("1. Add Flight")
    print("2. Find Flights")
    print("3. Book Seat")
    print("4. Cancel Reservation")
    print("5. Generate Report")
    print("6. Exit")
    choice = input("Enter your choice: ")
    if choice == "1":
        add_flight()
    elif choice == "2":
        find_flight()
    elif choice == "3":
        book_seat()
    elif choice == "4":
        cancel_reservation()
    elif choice == "5":
        generate_report()
    elif choice == "6":
        break
    else:
        print("Invalid choice. Please try again.")
```
